```python
import pandas as pd
import matplotlib.pyplot as plt
carros = pd.read_csv('precos_carros_brasil.csv', low_memory=False)
brand_counts = carros['brand'].value_counts()
topbrands = brand_counts.head(7)
plt.figure(figsize=(10, 8))
brand_counts.plot(kind='bar', color='darkred')


plt.title('Contagem de Marcas')
plt.ylabel('Quantidade')
plt.xlabel('Marcas')
plt.xticks(rotation=25, ha="right")
#plt.tight_layout()
plt.show()
```
