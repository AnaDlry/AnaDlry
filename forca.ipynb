```python
import random

def desenho(erros):
  if erros == 0:
      print()
      print("|----- ")
      print("|    | ")
      print("|      ")
      print("|      ")
      print("|      ")
      print("|      ")
      print("_      ")
      print()
  elif erros == 1:
      print()
      print("|----- ")
      print("|    | ")
      print("|    O ")
      print("|      ")
      print("|      ")
      print("|      ")
      print("_      ")
      print()
  elif erros == 2:
      print()
      print("|----- ")
      print("|    | ")
      print("|    O ")
      print("|    | ")
      print("|      ")
      print("|      ")
      print("_      ")
      print()
  elif erros == 3:
      print()
      print("|----- ")
      print("|    | ")
      print("|    O ")
      print("|   /| ")
      print("|      ")
      print("|      ")
      print("_      ")
      print()
  elif erros == 4:
      print()
      print("|----- ")
      print("|    | ")
      print("|    O ")
      print("|   /|\\ ")
      print("|      ")
      print("|      ")
      print("_      ")
      print()
  elif erros == 5:
      print()
      print("|----- ")
      print("|    | ")
      print("|    O ")
      print("|   /|\\ ")
      print("|   /  ")
      print("|      ")
      print("_      ")
      print()
  elif erros == 6:
      print()
      print("|----- ")
      print("|    | ")
      print("|    O ")
      print("|   /|\\ ")
      print("|    | ")
      print("|   / \\ ")
      print("_      ")
      print()

palavras = ['capacete', 'porcelana', 'combustivel', 'dardo', 'convexo', 'pedestre', 'noivo']
palavra_sorteada = random.choice(palavras)
letras = [letra for letra in palavra_sorteada]

print ("--- JOGO DA FORCA ---")

tentativas = []
erros = 0

def mostrar_palavra(): #treco pra printar os _ _ _
  for letra in palavra_sorteada:
    if letra in tentativas:
      print(letra, end=' ')
    else:
      print('_', end=' ')
  print()

while erros < 6:
  desenho(erros)
  print(f"\nErros: {erros}/6")
  print(f"\nLetras tentadas: {tentativas}")
  mostrar_palavra()

  chute = input('chute uma letra: ').lower()
  if not chute.isalpha() or len(chute) != 1: #se não for alfabeto ou se for mais de uma letra
    print("Digite apenas uma letra!")
    continue
  if chute in tentativas:
    print("Você já tentou essa letra!")
    continue
  tentativas.append(chute)

  if chute in letras:
    print ("Você acertou uma letra!")
  else:
    print("Não há essa letra na palavra!")
    erros +=1

  acertou = True  #definir se ganhou ou não
  for letra in palavra_sorteada:
    if letra not in tentativas:
      acertou = False
      break

  if acertou:
    print("\n===============================================")
    print("PARABÉNS! Você acertou a palavra: ")
    mostrar_palavra()
    print("===============================================")
    break

if erros == 6:
    print("\n=============================")
    print("ERROU! Você foi enforcado!")
    desenho(erros)
    print(f"A palavra era: {palavra_sorteada} ")
    print("=============================")


```
